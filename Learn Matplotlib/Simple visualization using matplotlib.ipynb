# Here's some of my python matplotlib exercise
Simple data visualization using matplotlib


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
kids = ['Ruben','Baim','Nagita','Ayu','Anang','Syahrini']
revenue_2018 = [50,60,50,43,70,-10]
revenue_2019 = [80,80,50,40,60,-50]
```

# Line Chart
label > each legend of line

color > color the line

marker > mark each point

markerfacecolor > color the marker name with rgb color

linestyle > line style there's (solid/dotted/dashed/dashdot)

```python
plt.plot(kids, revenue_2018, label = 'Revenue 2018',color='magenta',alpha=0.7, marker='o',markerfacecolor='red', linestyle ='dashed',linewidth=2)
plt.plot(kids, revenue_2019, label = 'Revenue 2019',color='teal',alpha=0.8, marker='o', linewidth=2)
plt.xlabel('Kids')
plt.ylabel('Revenue')
plt.title('Kids Revenue')
plt.legend(loc='lower left')
plt.show()
```

# Stacked Area Chart

```python
plt.fill_between(kids,  revenue_2018, color="lightpink", alpha=0.7, label='Revenue 2018')
plt.fill_between(kids,  revenue_2019, color="skyblue", alpha=0.7, label='Revenue 2019')


plt.xlabel('Kids')
plt.ylabel('Revenue')
plt.title('Kids Revenue')
plt.legend(loc='best')
plt.show()
```

# Scatter Chart

```python
plt.scatter(kids,revenue_2018,facecolor='magenta',alpha=0.7,label='Revenue 2018')
plt.scatter(kids,revenue_2019,facecolor='teal',alpha=0.8,label='Revenue 2019')
plt.xlabel('Kids')
plt.ylabel('Revenue')
plt.title('Kids Revenue')

plt.grid(linestyle='dotted',linewidth=0.9)

plt.legend(loc='best')
plt.show()
```

# Histogram Chart

```python
plt.hist(revenue_2018,3,facecolor='magenta', alpha=0.7,label='Revenue 2018')
plt.hist(revenue_2019,3,facecolor='teal',alpha=0.7,label='Revenue 2019')

plt.xlabel('Revenue')
plt.ylabel('Kids')
plt.legend(loc='best')
plt.show()
```

# Bar Chart

```python
N = 6
ind = np.arange(N) 
width = 0.35       
plt.bar(ind, revenue_2018, width, facecolor='magenta',alpha=0.5,label='Revenue 2018')
plt.bar(ind + width, revenue_2019, width,facecolor='teal',alpha=0.8,label='Revenue 2019')

plt.xlabel('Kids')
plt.ylabel('Revenue')
plt.title('Kids Revenue')

plt.xticks(ind + width / 2, (kids))
plt.legend(loc='upper right')
plt.show()
```
